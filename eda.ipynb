```python
df['Survived'] = df['Survived'].map({'no': 0, 'yes': 1})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

survival_by_gender = df.groupby('Sex')['Survived'].mean()

sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values,palette='viridis')

# Add titles and labels
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')

# Show the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

survival_by_gender = df.groupby('Pclass')['Survived'].mean()

sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values,palette='viridis')

# Add titles and labels
plt.title('Survival Rates by Pclass')
plt.xlabel('Class')
plt.ylabel('Survival Rate')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.scatterplot(x='Age', y='Fare', data=df, color='blue')

# Add titles and labels
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

```python
survival_by_gender_class = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(survival_by_gender_class, annot=True, cmap='coolwarm', fmt='.2f')

# Add titles and labels
plt.title('Survival Rates by Gender and Class')
plt.xlabel('Class')
plt.ylabel('Gender')

# Show the plot
plt.show()
```

```python
# Select only the numerical columns
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Add titles and labels
plt.title('Correlation Matrix of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Features')

# Show the plot
plt.show()
```

```python

# Extract part before the first comma
df['title'] = df['Name'].apply(lambda name: name.split(',')[0].strip().lower())



```

```python
df['Fare'].describe()
```

```python
import numpy as np

# Define conditions
conditions = [
    df['Fare'] <= 8,
    (df['Fare'] > 8) & (df['Fare'] < 15),
    df['Fare'] >= 15
]


values = ['low', 'mid', 'high']

df['Fare_range'] = np.select(conditions, values, default='unknown')

```

```python
df['Family_size']=df['Parch']+df['SibSp']+1
```

```python
sns.histplot(df['Family_size'], kde=True, bins=20, color='skyblue')

# Add titles and labels
plt.title('Histogram and KDE Plot Family Size')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

family_survival = df.groupby('Family_size')['Survived'].mean().reset_index()

# Plot the survival rate vs family size
plt.figure(figsize=(10, 6))
sns.barplot(x='Family_size', y='Survived', data=family_survival, palette='viridis')

# Add titles and labels
plt.title('Survival Rate vs Family Size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.xticks(ticks=range(family_survival['Family_size'].min(), family_survival['Family_size'].max() + 1))

# Show the plot
plt.show()
```
